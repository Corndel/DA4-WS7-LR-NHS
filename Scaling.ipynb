# Scaling data

```python
# libraries
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
#LOading in Data and visualising
df_dia = pd.read_csv("https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt",sep="\t")
df_dia.describe()
```

```python
# get column names to select which ones to scale
df_dia.columns
```

```python
# Applying standard scalar to selected columns
# This transforms data to have a mean of 0 and a SD of 1
# we have ignored the SEX columnn as we do not want to scale this
df_dia[['AGE', 'BMI', 'BP', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'Y']] =  StandardScaler().fit_transform(df_dia[['AGE', 'BMI', 'BP', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'Y']])
                                                                                                   
# Now this will work
df_dia.describe().round(2)
```
