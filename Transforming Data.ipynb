# Transforming features
Is a column is skewed we should transform this to be normally distributed.\
Linear models expected all imputs and outputs to be normally distributed\
The two most common ways are to take logs or roots\
Logs will not work if there are any 0 values, different methods yield diffeerent results

<img src="Data/Skew.png" alt="Corndel" width ="600" align="centre">

```python
import pandas as pd

import numpy as np
```

```python
# Import CSV and view distributions
df_dia = pd.read_csv("https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt",sep="\t")
df_dia.hist(figsize=(20,20), bins =50);
```

Candidates for transfroming are BMI, S2, S3 and Y

# Examine the output y

```python
# Hist of y
df_dia['Y'].hist(bins=50);
```

```python
# Data is not normally distributed
```

```python
# Applying a log transform
np.log(df_dia['Y']).hist(bins=50);
```

### Hasn't really helped, try taking roots

```python
# Taking roots
np.sqrt(df_dia['Y']).hist(bins=50);
```

### It is now more suitable for the model

```python
# Applying transformation
df_dia['Y'] = np.sqrt(df_dia['Y'])
```

```python
df_dia.hist(figsize=(20,20), bins =50);
```
